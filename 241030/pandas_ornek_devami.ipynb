```python
# Çalışanlar veri seti
emp = pd.DataFrame({
    'Employee ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department ID': [101, 102, 101, 103, 102],
    'Salary (USD)': [60000, 50000, 70000, 48000, 52000]
})
```

```python
# Departmanlar veri seti
dep = pd.DataFrame({
    'Department ID': [101, 102, 103],
    'Department Name': ['Marketing', 'Sales', 'HR'],
    'Location': ['New York', 'Los Angeles', 'Chicago']
})
```

```python
emp.columns=["Çalışan ID","İsim","Departman ID","Maaş"]
emp
```

```python
dep.columns=["Departman ID","Departman","Konum"]
dep
```

```python
#concat ile birleştirme.
df=pd.concat([emp,dep],ignore_index=True)
df
```

```python
#merge ile birleştirme.
df=pd.merge(emp,dep, on="Departman ID")
df
```

```python
#join ile birleştirme.
#ortak bir index belitmemiz gerekiyor.
dep.set_index("Departman ID",inplace=True)
emp.set_index("Departman ID",inplace=True)
df=emp.join(dep)
df
```
